# Linear Regression using only single feature

```python
import pandas as pd
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df
```

### creating a non smoker dataframe

```python
no_smoke_df=df[df.smoker=="no"]
```

```python
plt.title("AGE vs CHARGES")
sns.scatterplot(data=no_smoke_df,x='age',y='charges',alpha=0.7,s=15)
```

### charges= w x age + b  
### linear regression

```python
def estimate_charges(age,w,b):
    return age*w+b
```

```python
w=50
b=100
```

### creating a ages list adding all the ages of the non smokers and and created a estimated_charges as list with with vales from the function and ploting them

```python
ages=no_smoke_df.age
estimated_charges=estimate_charges(ages,w,b)
```

```python
plt.plot(ages,estimated_charges,'b-')
plt.xlabel('age')
plt.ylabel('charges')
```

### checking the estimated with the actual values

```python
target=no_smoke_df.charges
plt.plot(ages,estimated_charges,'r',alpha=0.8)
plt.scatter(ages,target,s=8,alpha=0.8)
plt.xlabel('age')
plt.ylabel('charger')
plt.legend(['estimate','actual'])
```

### creating a function by adding only weight and bias it will be giving the plot

```python
def try_variable(w,b):
    ages=no_smoke_df.age
    target=no_smoke_df.charges
    estimated_charges=estimate_charges(ages,w,b)
    plt.plot(ages,estimated_charges,'r',alpha=0.8)
    plt.scatter(ages,target,s=8,alpha=0.8)
    plt.xlabel('age')
    plt.ylabel('charges')
    plt.legend(['estimated','actual'])
```

```python
try_variable(300,-4500)
```

```python
!pip install numpy --quiet
```

```python
import numpy as np
```

```python
target=no_smoke_df.charges
predicted=estimated_charges
def rmse(target,predicted):
    return np.sqrt(np.mean(np.square(target-predicted)))
rmse(target,predicted)
```

```python
def try_variable(w,b):
    ages=no_smoke_df.age
    target=no_smoke_df.charges
    estimated_charges=estimate_charges(ages,w,b)
    plt.plot(ages,estimated_charges,'r',alpha=0.8)
    plt.scatter(ages,target,s=8,alpha=0.8)
    plt.xlabel('age')
    plt.ylabel('charges')
    plt.legend(['estimated','actual'])
    predicted=estimated_charges
    print("loss: " ,rmse(target,predicted))
    
```

```python
try_variable(267.24891283,-2091.4205565650864)
```

# normally the numerics are having large difference so we have standardize we get a optimal weight

```python
from sklearn.preprocessing import StandardScaler
```

```python
numeric_col=['age','bmi','children']
scaler=StandardScaler()
scaler.fit(df[numeric_col])
```

```python
scaler.mean_
```

```python
scaler.var_
```

```python
scaler_inputs=scaler.transform(df[numeric_col])
scaler_inputs
```

```python
cat_col=['smoker_num','sex_num','northeast', 'northwest', 'southeast', 'southwest']
catagorical_data=df[cat_col].values
```

```python
inputs=np.concatenate((scaler_inputs,catagorical_data),axis=1)
target=df.charges
model=LinearRegression()
model.fit(inputs,target)
prediction=model.predict(inputs)
rmse(prediction,target)
```

```python
weight=pd.DataFrame({
    'fearture':np.append(numeric_col+cat_col,'bias'),
    'values':np.append(model.coef_,model.intercept_)
})
weight
```

# main thing is creating a train test data and calculating loss

```python
from sklearn.model_selection import train_test_split
```

```python
input_train,input_test,target_train,target_test=train_test_split(inputs,target,test_size=0.10,random_state=42)
```

```python
model.fit(input_train,target_train)
prediction=model.predict(input_test)
rmse(prediction,target_test)
```

```python
pre_train=model.predict(input_train)
rmse(pre_train,target_train)
```
